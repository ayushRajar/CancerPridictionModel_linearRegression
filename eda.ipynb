<h3> outlier detection </h3>

```python
processed_data=pd.read_csv('Cancer_prediction_dataset/cancer_reg_process.csv')
processed_data.head()
```

```python
from scipy.stats import normaltest
numerical_columns =processed_data.select_dtypes(include=np.number).columns
gaussian_cols = []
non_gaussian_cols = []

for col in numerical_columns:
    stat,p = normaltest(processed_data[col])
    print('Statistics=%.3f, p=%.3f' %(stat,p))
    alpha = 0.05
    if(p > alpha):
        gaussian_cols.append(col)
    else:
        non_gaussian_cols.append(col)
print(gaussian_cols)
    
```
